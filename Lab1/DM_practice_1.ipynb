# IV. Data Cleaning

11. (3б) Перетворити колонку launched у цілочислену (Int64). Перевірити, чи все пройшло, як очікувалось.

```python
missile_dataframe['launched'] = missile_dataframe['launched'].astype("Int64")
```

```python
missile_dataframe.info()
```

12. (3б) Видалити колонки, де кількість непустих значень становить менше 30% від загальної кількості записів. Можна вручну порахувати і тут тільки видалити. За автоматизацію розрахунку відстоків - додаткові 3 бали.

```python
percent = missile_dataframe.isna().sum() / len(missile_dataframe) * 100
print(percent, '\n')

missile_dataframe = missile_dataframe.drop(["launch_place", 'destroyed_details', 'carrier', 'not_reach_goal', 'source'], axis=1)
# also source
missile_dataframe.info()
```

13. (4б) Очистити датафрейм від дублікатів (дублюючихся рядків) та записів, де атака продовжувалась відʼємну або нульову кількість часу (що фізично неможливо). Не забудьте оновити індекс. Скільки рядків було видалено? Як ви це перевірили у коді?

```python
rows_before = len(missile_dataframe)
print(f"rows_before {rows_before}")
missile_dataframe = missile_dataframe.drop_duplicates()

missile_dataframe['time_start'] = missile_dataframe['time_start'].apply(lambda x: x if len(x) > 11 else x + ' 00:00')
missile_dataframe['time_end'] = missile_dataframe['time_end'].apply(lambda x: x if len(x) > 11 else x + ' 00:00')

missile_dataframe['time_start'] = pd.to_datetime(missile_dataframe['time_start'], format='%Y-%m-%d %H:%M')
missile_dataframe['time_end'] = pd.to_datetime(missile_dataframe['time_end'], format='%Y-%m-%d %H:%M')
missile_dataframe = missile_dataframe[missile_dataframe['time_end'].dt.time >= missile_dataframe['time_start'].dt.time]

missile_dataframe.reset_index(drop=True, inplace=True)

rows_after = len(missile_dataframe)
print(f"rows_after {rows_after}")
```

```python
unique_year = missile_dataframe['time_start'].dt.year.unique()
duration_list = []
for year in unique_year:
    unique_months = missile_dataframe[missile_dataframe['time_start'].dt.year == year]['time_start'].dt.month.unique()
    for month in unique_months:
        filter = missile_dataframe[(missile_dataframe['time_start'].dt.year == year) & (missile_dataframe['time_end'].dt.month == month)]
        max_hour = (filter['time_end'] - filter['time_start']).max()
        duration_list.append({'year': year, 'month': month, 'attack_duration': max_hour})

duration_dataframe = pd.DataFrame(duration_list).to_csv('max_duration.csv', index=False)

duration_read = pd.read_csv('max_duration.csv')
duration_read
```
